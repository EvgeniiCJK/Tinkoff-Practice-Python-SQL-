## Строим Диаграмму с накоплениями Верификаций ЛАЙТ

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime, time

# создаём папочку для хранения построенных диаграмм
dir = '/mnt/zeppelin-exchange/e.s.kopylov/vrf-672/'
if not os.path.exists(dir):
  os.makedirs(dir)
try:
    # Размечаем цвета 
    mycolors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink']      
    # Размечаем оси
    fig, ax = plt.subplots(1,1,figsize=(28, 9), dpi= 80)
    columns = VRF_672_1.columns[1:]
    labs = columns.values.tolist()
    # Закинем данные в список
    R = VRF_672_1.groupby('дата').mean().reset_index()
    x = range(len(R['дата'])) # индексируем ось x
    ax.set_xticks(x)
    ax.set_xticklabels(R['дата'], rotation = 90) # устанавливаем фио по оси Ох 
    y0 = VRF_672_1[columns[0]].values.tolist()
    y1 = VRF_672_1[columns[1]].values.tolist()
    y2 = VRF_672_1[columns[2]].values.tolist()
    y3 = VRF_672_1[columns[3]].values.tolist()
    y4 = VRF_672_1[columns[4]].values.tolist()
    y5 = VRF_672_1[columns[5]].values.tolist()
    # соединим массивы по вертикали с помощью vstack
    y = np.vstack([y0, y1, y2, y3, y4, y5])
    # строим диаграмму для каждой колонки
    labs = columns.values.tolist()
    ax = plt.gca()
    ax.stackplot(x, y, labels=labs, colors=mycolors, alpha=0.8)
    # обзываем нашу диаграмму
    ax.set_title('Диаграмма с накоплениями Верификаций ЛАЙТ', fontsize=24)
    ax.set(ylim=[0, 100])
    # устанавливаем размер легенды и расположение
    ax.legend(fontsize=10, ncol=3, loc='upper center')
    # устанавливаем интервал по x
    plt.xticks(x[::1], fontsize=12, horizontalalignment='center')
    # устанавливаем интервал по y
    plt.yticks(np.arange(0, 2500, 100), fontsize=10)
    plt.xlabel('Дата', fontsize = 16)
    plt.ylabel('Количество заявок', fontsize = 16)
    plt.xlim(x[0], x[-1])
    # gca - даёт ссылки на текущие оси, spines - для плавного соединения точек
    plt.gca().spines["top"].set_alpha(0)
    plt.gca().spines["bottom"].set_alpha(.3)
    plt.gca().spines["right"].set_alpha(0)
    plt.gca().spines["left"].set_alpha(.3)
    filename3 = 'Диаграмма с накоплениями Верификаций ЛАЙТ.png' # указываем путь для сохранения
    img3 = dir + filename3
    plt.savefig(img3, bbox_inches='tight') # сохраняем полученную диаграмму в сжатом виде
    plt.show()
    print("Удачно!")
except:
    print("Ошибка! Что-то пошло не так.")
```
